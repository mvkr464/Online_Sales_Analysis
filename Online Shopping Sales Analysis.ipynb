# Exploratory Data Analysis

```python
# Showing count of Gender in Bar graph 
```

```python
df.columns
```

```python
G = sns.countplot(x = 'Gender', data = df)
for bars in G.containers:
    G.bar_label(bars)
```

```python
# Sum of sales amount by Gender wise
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values('Amount')
```

```python
# Sum of sales by Gender wise in Barplot
ts = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values('Amount')
ts = sns.barplot(x = 'Gender', y = 'Amount', data = ts);
for bars in ts.containers:
    ts.bar_label(bars)
```

```python
# Age group wise the no of sales shown in countplot
palette = ['pink', 'blue']
ag = sns.countplot(x = 'Age Group', hue = 'Gender', data = df, palette = palette)
for bars in ag.containers:
    ag.bar_label(bars)
```

```python
# Sum of sales by Age group wise in Barplot
age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values('Amount', ascending = False)
sns.set(rc = {'figure.figsize':(10,5)})
age = sns.barplot(x = 'Age Group', y = 'Amount', data = age);
for bars in age.containers:
    age.bar_label(bars)
```

```python
# Sum of sales by State wise in Barplot for top 10
state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values('Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,7)})
state = sns.barplot(x = 'State', y = 'Amount', data = state);
for bars in state.containers:
    state.bar_label(bars)
```

```python
# Sum of orders by State wise in Barplot for top 10
state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values('Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,7)})
state = sns.barplot(x = 'State', y = 'Orders', data = state);
for bars in state.containers:
    state.bar_label(bars)
```

```python
# Showing the martial status vs no of orders placed
sns.set(rc = {'figure.figsize':(5,7)})
ma = sns.countplot(x = 'Marital_Status', data = df)
for bars in ma.containers:
    ma.bar_label(bars)
```

```python
# Sum of sales by Gender wise along with marital status in Barplot
mf = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values('Amount', ascending = False)
sns.set(rc = {'figure.figsize':(10,7)})
mf = sns.barplot(x = 'Marital_Status', y = 'Amount', data = mf, hue = 'Gender');
for bars in mf.containers:
    mf.bar_label(bars)
```

```python
# No of orders placed as per occupation wise
sns.set(rc = {'figure.figsize':(20,7)})
oc = df.sort_values('Occupation',ascending = False)
oc = sns.countplot(x = 'Occupation', data = df)
for bars in oc.containers:
    oc.bar_label(bars)
```

```python
# Sum of sales by Occupation wise in Barplot
occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values('Amount', ascending = False)
sns.set(rc = {'figure.figsize':(20,7)})
occ = sns.barplot(x = 'Occupation', y = 'Amount', data = occ);
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
# no of orders by product category wise
pc = sns.countplot(x = 'Product_Category', data = df)
for bars in pc.containers:
    pc.bar_label(bars)
```

```python
# Sum of sales by Product Category wise in Barplot
occu = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending = False)
sns.set(rc = {'figure.figsize':(25,5)})
occu = sns.barplot(x = 'Product_Category', y = 'Amount', data = occu);
for bars in occu.containers:
    occu.bar_label(bars)
```
